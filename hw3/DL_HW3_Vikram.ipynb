```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_excel("/content/extra_credits_dataset.xlsx", "Sheet1")
```

```python
# Solution for question 1 and 2
df.describe()
```

```python
# Solution for question 3
(df[df["Extra"] <= 1].count() / df[df["Extra"] > 0].count()) * 100
```

```python
# Solution for question 4
df.corr()
```

```python
# Solution for question 5
plt.scatter(df["Extra"], df["Total"])
plt.xlabel("Extra Credits")
plt.ylabel("Final Score")
plt.show()
```

```python
# Solution for question 6
from sklearn import linear_model
df = df[df["Extra"] > 0]
X = df[['Extra']]
Y = df[['Total']]
reg = linear_model.LinearRegression()
reg.fit(X, Y)
```

```python
print("Coefficient", reg.coef_)
print("Intercept", reg.intercept_)
print("Therefore, linear equation is :", reg.coef_, "*(extra marks) +", reg.intercept_)
```

```python
# Solution for question 7
from sklearn import metrics
import math
y_predict = reg.predict(X)
print(math.sqrt(metrics.mean_squared_error(Y, y_predict)))
```

```python
# Solution for question 8
plt.scatter(X, Y, color = 'blue')
plt.plot(X, reg.predict(X), color = 'red')
plt.xlabel('extra credit')
plt.ylabel('final marks')
plt.show()
```
